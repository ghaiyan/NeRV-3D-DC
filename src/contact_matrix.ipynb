```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

rawmatrix=np.loadtxt("../chr19_50kb/chr19_50000_raw.hic")

normalized=np.loadtxt("../chr19_50kb/chr19_50000.hic")

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
sns.set_theme(style="white")
sns.heatmap(rawmatrix[::2,::2], robust=True, cmap="Reds", xticklabels=False, yticklabels=False)
plt.title('raw contact matrix')

plt.subplot(1,2,2)
sns.heatmap(normalized[::2,::2], robust=True, cmap="Reds")
plt.title('normalized contact matrix')

plt.show()
```
